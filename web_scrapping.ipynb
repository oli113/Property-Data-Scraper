```python
import requests
from bs4 import BeautifulSoup

def get_property_details(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')

    details = {}
    features_div = soup.find('div', 'expandable-box__hidden js-expandable-box-target')
    if features_div:
        features_list = features_div.find_all('li')
        for feature in features_list:
            label = feature.find('span', class_='label').get_text(strip=True)
            value = feature.find('span', class_='value').get_text(strip=True)
            details[label] = value
    return details

data_list = []
BASE_URL = "https://www.royallepage.ca/en/search/homes/{page}/?property_type=&house_type=&features=&listing_type=&lat=42.988148&lng=-81.246092&bypass=&address=London&address_type=city&city_name=London&prov_code=ON&display_type=gallery-view&da_id=&travel_time=false&school_id=&search_str=London%2C+ON%2C+CAN&id_search_str=UK&school_search_str=&travel_time_min=30&travel_time_mode=drive&travel_time_congestion=&min_price=0&max_price=5000000%2B&min_leaseprice=0&max_leaseprice=5000%2B&beds=0&baths=0&transactionType=SALE&keyword="
# Loop through pages 1 to 10
for page in range(1, 18):
    url = BASE_URL.format(page=page)
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'lxml')
    
    listing_divs = soup.find_all('div', class_='card card--listing-card js-listing js-property-details')
    
    for div in listing_divs:
        data = {}
        
        # Extract link
        link_tag = div.find('a')
        link = link_tag['href'] if link_tag else None

        # Extract price
        price_tag = div.find('span', class_='title--h3 price')
        price = price_tag.get_text(strip=True) if price_tag else None

        data['Link'] = link
        data['Price'] = price

        # Get property details and add them to the data dictionary
        if link:
            details = get_property_details(link)
            data.update(details)

        data_list.append(data)

print(data_list)
```

```python
import pandas as pd
df = pd.DataFrame(data_list)

# Save the DataFrame to a CSV file (optional)
df.to_csv('property_data.csv', index=False)

# Print the DataFrame to see the data
print(df)
```
