```python
#As there were so many column extracted from website. 
# And there are column in which most of the data missing. 
# I will delete the columns which has missing value more then 40%

def drop_columns_with_missing_values(df, threshold=0.4):
    missing_proportions = df.isnull().mean()
    cols_to_drop = missing_proportions[missing_proportions > threshold].index
    df.drop(columns=cols_to_drop, inplace=True)
    return df

```

```python
# 1. price need to be int and $ sign need to be remove
import pandas as pd

def convert_price_to_number(df, column_name):
 
    # Remove $ sign and commas, and convert to integer
    df[column_name] = df[column_name].str.replace('$', '').str.replace(',', '').astype(float)
    return df


data = convert_price_to_number(data, 'Price')
print(data)
```

```python
# 2. bedrooms, bathrooms column name, datatype need to change 
# There are some value 4+1 it also need to be change

def convert_to_int(val):
    res = []
    for i in val:
        if pd.isna(i):
            res.append(i)
            continue
        str_val = str(i)
        if '+' in str_val:
            s = sum(int(num) for num in str_val.split('+'))
            res.append(s)
        else:
            res.append(int(float(str_val)))  
    return res
    # try:
    #     return int(eval(val))
    # except:
    #     return val

```

```python
# data['Bathrooms:'] = data['Bathrooms:'].apply(convert_to_int)
data.rename(columns={'Bathrooms:': 'NumBathrooms'}, inplace=True)
bathrooms_value =  convert_to_int(data['NumBathrooms'])
data['NumBathrooms'] = bathrooms_value

data.head()
```

```python
data.rename(columns={'Bedrooms:': 'NumBedrooms'}, inplace=True)
bedroom_value =  convert_to_int(data['NumBedrooms'])
data['NumBedrooms'] = bedroom_value

data.head()
```

```python
def check_bathroom_level(bathroom_value):
    bathroom_level = []
    levels = ['Main Level', 'Other level', '2nd Level','3rd Level','Lower Level', 'Upper Level', 'Ground Level', 'Basement Level']
    for i in bathroom_value:
        found = False
        
        for level in levels:
            if level in str(i):
                bathroom_level.append(level)
                found = True
                break

        if not found:
            bathroom_level.append(i)

        
    return bathroom_level
        


 
```
